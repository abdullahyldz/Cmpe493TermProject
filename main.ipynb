### train_data is dictionary of dictionary
<br>
first key is file name like BB-norm-10658649, second key is extention like .txt

```python
def readData(folder):
    directories = os.listdir(folder)
    if('LICENSE' in directories):
        directories.remove('LICENSE')
        print("removed licence")
    if('README' in directories):
        directories.remove('README')

    train_data = {}
    for directory in directories:
        regex = re.search('(.*?)\.(\w+)', directory)
        print(directory, ' -> ', regex[1], ' + ', regex[2], '\n')

        info = open(f'{folder}/{directory}', 'r', encoding="utf8").read()
        #print(info)

        if(directory not in train_data):
            train_data[regex[1]] = {}
        train_data[regex[1]][regex[2]] = info
    return train_data
```

```python
def whatIs(id):
    term = go[f'OBT:{id}']
    is_a = set(term.relationships.get(go['is_a']))
    # TODO Find synonyms for each equivalent term. 
    return is_a
```

```python
def getChildrenById(id):
    subnodes = {(term.id, term.name) for term in go[f"OBT:{id}"].subclasses(distance = 2, with_self=False)}
    return subnodes

getChildrenById('002488')
#  '002488' human
#  '003188' child
#  getChildren('003188') returns {'boy', 'child', 'girl'}

#       |
#     child
#     /    \
#   boy   girl
  
# 3188 -> child
# 3248 -> boy
# 3259 -> girl
```

```python
def getParentsById(id):
    supernodes = {(term.id, term.name) for term in go[f"OBT:{id}"].superclasses(distance=2, with_self=False)}
    return supernodes

getParentsById('002488')
```

```python
def getOntologyPairs():
    pairs = [(term.id, term.name) for term in go.terms()]
    return pairs
```

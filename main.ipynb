### train_data is dictionary of dictionary
<br>
first key is file name like BB-norm-10658649, second key is extention like .txt

```python
def readData(folder):
    directories = os.listdir(folder)
    if('LICENSE' in directories):
        directories.remove('LICENSE')
        print("removed licence")
    if('README' in directories):
        directories.remove('README')

    train_data = {}
    for directory in directories:
        regex = re.search('(.*?)\.(\w+)', directory)
        print(directory, ' -> ', regex[1], ' + ', regex[2], '\n')

        info = open(f'{folder}/{directory}', 'r', encoding="utf8").read()
        #print(info)

        if(directory not in train_data):
            train_data[regex[1]] = {}
        train_data[regex[1]][regex[2]] = info
    return train_data
```

```python
train_data = readData('./train')
```
